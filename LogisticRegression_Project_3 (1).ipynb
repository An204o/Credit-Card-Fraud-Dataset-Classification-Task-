```python
import pandas as pd

#displaying the dataset with first 5 rows
df = pd.read_csv("creditcard.csv").drop(columns = 'Time', axis = 1)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

#scaling the data for all columns to make std less, except the Class column
df_scaled = StandardScaler().fit_transform(df.drop("Class", axis = 1))
```

```python
#We are converting df_scaled to dataframe and include the class column
df_scaled = pd.DataFrame(df_scaled, columns = df.drop("Class", axis = 1).columns)
df_scaled["Class"] = df["Class"].reset_index(drop=True)
df_scaled.head()
```

```python
#the table shows that the values have been scaled, which means mean = 0 and std = 1
df_scaled.describe()
```

```python
#In this part, we will be finding how the independent variables have influence or effect on a dependet variable(Class feature)

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split


lr = LogisticRegression()
X = df_scaled.drop("Class", axis = 1)
y = df_scaled["Class"]
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42,
    stratify=y
    )

lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)

```

```python
#we will check the trained model accuracy

from sklearn.metrics import f1_score

x_pred = lr.predict(X_train)
f1_score(y_train, x_pred)
```

```python
#we will use the accuracy score to verify how many correct predictions are made out of the total predictions by using the f1-score

from sklearn.metrics import f1_score

y_pred = lr.predict(X_test)
f1_score(y_test, y_pred)
```

```python
#Finally, I will use the confusion matrix to display the correctly and incorrectly predicted number of postive and negatives compared to the actual values.
#The model predicts most non-fraud transactions correctly (TN = 85280) and catches a good portion of fraud cases (TP = 91).

from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(lr, X_test, y_test)
```
